# <span style="color: rgb(34, 34, 34); font-family: &quot;.SFUI-Regular&quot;; font-size: 12px; background-color: rgb(255, 255, 255);">Вывести список товаров ценой больше 20 долларов</span>

```sql
SELECT *
  FROM [Production].[Products] 
  WHERE [unitprice] > 20
```

# <span style="color: rgb(34, 34, 34); font-family: &quot;.SFUI-Regular&quot;; font-size: 12px; background-color: rgb(255, 255, 255);">Вывести список поставщиков, в данных которых не указан регион</span>

```sql
SELECT [custid]
      ,[companyname]
      ,[contactname]
      ,[contacttitle]
      ,[address]
      ,[city]
      ,[region]
      ,[postalcode]
      ,[country]
      ,[phone]
      ,[fax]
  FROM [Sales].[Customers]
  WHERE [region] IS NULL;
```

#  <span style="color: rgb(34, 34, 34); font-family: &quot;.SFUI-Regular&quot;; font-size: 12px; background-color: rgb(255, 255, 255);">&nbsp;Выбрать 10 самых дешевых (по цене за единицу) товаров, которые имеются в продаже</span>

```sql
SELECT TOP 10 [productid]
      ,[productname]
      ,[supplierid]
      ,[categoryid]
      ,[unitprice]
      ,[discontinued]
  FROM [Production].[Products]
  ORDER BY [unitprice];
```

# <span style="color: rgb(34, 34, 34); font-family: &quot;.SFUI-Regular&quot;; font-size: 12px; background-color: rgb(255, 255, 255);">Вывести список категорий товаров, которые были проданы в 2006 году</span>

```sql
SELECT DISTINCT [Production].[Categories].[categoryid]
      ,[Production].[Categories].[categoryname]
      ,[Sales].[Orders].[год]
  FROM [Production].[Categories]    
  JOIN [Production].[Products] ON [Production].[Categories].[categoryid] = [Production].[Products].[categoryid]
  JOIN [Sales].[OrderDetails] ON [Production].[Products].[productid] = [Sales].[OrderDetails].[productid]
  JOIN [Sales].[Orders] ON [Sales].[OrderDetails].[orderid] = [Sales].[Orders].[orderid]
  WHERE [Sales].[Orders].[год] = '2006';
```

# <span style="color: rgb(34, 34, 34); font-family: &quot;.SFUI-Regular&quot;; font-size: 12px; background-color: rgb(255, 255, 255);">Вывести для каждого клиента информацию о его/ее первом заказе</span>

<span style="color: rgb(34, 34, 34); font-family: &quot;.SFUI-Regular&quot;; font-size: 12px; background-color: rgb(255, 255, 255);"><br></span>

```sql
 WITH CTE AS 
(SELECT [orderid]
      ,[custid]
      ,[empid]
      ,[orderdate]
      ,[requireddate]
      ,[shippeddate]
      ,[shipperid]
      ,[freight]
      ,[shipname]
      ,[shipaddress]
      ,[shipcity]
      ,[shipregion]
      ,[shippostalcode]
      ,[shipcountry]
      ,[год]
      , ROW_NUMBER()OVER(PARTITION BY [custid] ORDER BY [orderdate]) as num
  FROM [Sales].[Orders])
  SELECT *
    , [Sales].[Customers].[companyname]
    FROM CTE
    JOIN [Sales].[Customers] ON CTE.[custid] = [Sales].[Customers].[custid]
  WHERE [num] = 1;
```
